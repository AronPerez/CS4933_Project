## Train data

````python
# https://stackoverflow.com/questions/48899051/how-to-drop-a-specific-column-of-csv-file-while-reading-it-using-pandas
# or
# https://stackoverflow.com/questions/24366449/python-pandas-how-to-skip-columns-when-reading-a-file/24366664

# Plot 2 Discriminator Loss
train_discriminator_Loss = pd.read_csv('/content/drive/MyDrive/CV/checkpoints/50epoch/log_train.dat',
            index_col=0, header=None, sep='\s',usecols=[0,2], engine='python')
# Plot 3 Discriminator Loss Fake
train_discriminator_Loss_Fake = pd.read_csv('/content/drive/MyDrive/CV/checkpoints/50epoch/log_train.dat',
            index_col=0, header=None, sep='\s',usecols=[0,3], engine='python')
# Plot 4 Discriminator Loss Real
train_discriminator_Loss_Real = pd.read_csv('/content/drive/MyDrive/CV/checkpoints/50epoch/log_train.dat',
            index_col=0, header=None, sep='\s',usecols=[0,4], engine='python')
# Plot 5 Generator Loss
train_generator_Loss = pd.read_csv('/content/drive/MyDrive/CV/checkpoints/50epoch/log_train.dat',
            index_col=0, header=None, sep='\s',usecols=[0,5], engine='python')
# Plot 6 Generator loss L1
train_generator_Loss_L1 = pd.read_csv('/content/drive/MyDrive/CV/checkpoints/50epoch/log_train.dat',
            index_col=0, header=None, sep='\s',usecols=[0,6], engine='python')
# Plot 7 Generator loss of GAN
train_generator_Loss_GAN = pd.read_csv('/content/drive/MyDrive/CV/checkpoints/50epoch/log_train.dat',
            index_col=0, header=None, sep='\s',usecols=[0,7], engine='python')
# Plot 8 Accuracy
train_accuracy = pd.read_csv('/content/drive/MyDrive/CV/checkpoints/50epoch/log_train.dat',
            index_col=0, header=None, sep='\s',usecols=[0,8], engine='python')

# https://stackoverflow.com/questions/42151460/pandas-dataframe-calculate-average-for-a-column-for-each-unique-index-without
#print(discriminator_Loss_Fake)
#print(discriminator_Loss_Fake.groupby(level=0).mean())
````

````python
fig, ax = plt.subplots()

ax.plot(train_discriminator_Loss.groupby(level=0).mean()) 

ax.set_title('Train Discriminator Loss')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
plt.show()
plt.draw()
plt.pause(.01)
````

````python
fig, ax = plt.subplots()

#ax.plot(moving_average(train_data[:,3], window_width)) 
ax.plot(discriminator_Loss_Fake.groupby(level=0).mean())

ax.set_title('Train Discriminator Loss Fake')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
plt.show()
plt.draw()
plt.pause(.01)
````

````python
fig, ax = plt.subplots()

ax.plot(train_discriminator_Loss_Real.groupby(level=0).mean()) 

ax.set_title('Train Discriminator Loss Real')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
plt.show()
plt.draw()
plt.pause(.01)
````

````python
fig, ax = plt.subplots()

ax.plot(train_generator_Loss.groupby(level=0).mean()) 

ax.set_title('Train Generator Loss')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
plt.show()
plt.draw()
plt.pause(.01)
````

````python
fig, ax = plt.subplots()

ax.plot(train_generator_Loss_L1.groupby(level=0).mean())

ax.set_title('Train Generator loss L1')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
plt.show()
plt.draw()
plt.pause(.01)
````

````python
fig, ax = plt.subplots()

ax.plot(train_generator_Loss_GAN.groupby(level=0).mean()) 

ax.set_title('Train Generator loss of GAN')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
plt.show()
plt.draw()
plt.pause(.01)
````

````python
fig, ax = plt.subplots()

ax.plot(train_accuracy.groupby(level=0).mean()) 

ax.set_title('Train Accuracy')
ax.set_ylabel('Accuracy of guess')
ax.set_xlabel('Epoch')
plt.show()
plt.draw()
plt.pause(.01)
````
